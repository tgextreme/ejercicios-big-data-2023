```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
response = requests.get('https://www.imdb.com/chart/moviemeter/')
soup = BeautifulSoup(response.text, 'html.parser')
movie_table = soup.find('table', {'class': 'chart'})
rows = movie_table.find_all('tr')
movies = []
for row in rows[1:]:
    title = row.find('td', {'class': 'titleColumn'}).a.text
    year = row.find('span', {'class': 'secondaryInfo'}).text.strip('()')
    rating = row.find('td', {'class': 'imdbRating'})
    print(rating)
    link = 'https://www.imdb.com' + row.find('td', {'class': 'titleColumn'}).a['href']
    movies.append({'title': title, 'year': year, 'rating': rating, 'link': link})
df = pd.DataFrame(movies)
print(df)
df.to_csv('imdb_movies.csv', index=False)
```
